```
import requests
import json
import pandas as pd
import requests
from datetime import datetime
```

```
parameters = {"q":"language:jupyter-notebook", "sort":"stars", "order":"desc", "per_page":100, "page":1}
results = requests.get('https://api.github.com/search/repositories?', params=parameters)
status_code = results.status_code
print(status_code)
```

```
data = results.json()
```

```
df = pd.DataFrame(columns=['reponame', 'URL', 'owner', 'repoID', 'stars', 'description'])
```

```
for repo in data['items']:
    temp_data = {
        'reponame':repo['name'],
        'URL':repo['html_url'],
        'owner':repo['owner']['login'],
        'repoID':repo['id'],
        'stars':repo['stargazers_count'],
        'description':repo['description']}

    df = df.append(temp_data, ignore_index=True)
```

```
df
```

```
udacity_top100 = df[df["owner"] == "udacity"]
udacity_top100
```
